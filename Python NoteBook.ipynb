<h3>Reading SO2 data of Delhi from the appropriate csv file and storing it in a pandas dataframe. 
<h3> We also create 'City', 'Month', 'Year' and 'Day' and store corresponding value in the dataframe.

```python
data1=pd.read_csv("GEE DATASETS USED\SO2_Sentinel\SO21\SO2Delhi.csv")
data1['City'] = 'Delhi'
data1.rename(columns = {'mean':'Mean'}, inplace = True) 
data1['Month'] = [month[int(x[3:5]) -1] for x in data1['date']]
data1['Year'] = [x[6:10] for x in data1['date']]

temp= pd.to_datetime(data1['date'])
temp = temp.dt.strftime('%d-%m-%Y')
data1['Day'] = pd.to_datetime(temp).dt.day_name()



data1
```

```python
Pollution = Pollution.reindex(columns=['City','date','Day','Month','Year','Mean Pollution'])
Pollution.rename(columns = {'date':'Date'}, inplace = True) 
```

<h3>Averaging the Pollution levels of all the wards on a particular date</h3>

```python
Pol_Del=Delhi_Pollution.groupby('date').mean()
Pol_Mum=Mumbai_Pollution.groupby('date').mean()
Pol_Kol=Kolkata_Pollution.groupby('date').mean()
Pol_Ban=Bangalore_Pollution.groupby('date').mean()
Pol_Hyd=Hyderabad_Pollution.groupby('date').mean()

dict = {'City':[], 'Date':[], 'Mean Pollution':[]}

for d in Delhi_Pollution.date.unique():
    dict['City'].append('Delhi')
    dict['Date'].append(d)
    dict['Mean Pollution'].append(Pol_Del['Mean Pollution'][d])
    
for d in Bangalore_Pollution.date.unique():
    dict['City'].append('Bangalore')
    dict['Date'].append(d)
    dict['Mean Pollution'].append(Pol_Ban['Mean Pollution'][d])

for d in Hyderabad_SO2.date.unique():
    if d in Pol_Hyd['Mean Pollution']:
        dict['Mean Pollution'].append(Pol_Hyd['Mean Pollution'][d])    
        dict['City'].append('Hyderabad')
        dict['Date'].append(d)
    
    
for d in Kolkata_Pollution.date.unique():
    dict['City'].append('Kolkata')
    dict['Date'].append(d)
    dict['Mean Pollution'].append(Pol_Kol['Mean Pollution'][d])

    
for d in Mumbai_Pollution.date.unique():
    dict['City'].append('Mumbai')
    dict['Date'].append(d)
    dict['Mean Pollution'].append(Pol_Mum['Mean Pollution'][d])

    

```

<h3>Creating the final dataset of Mean SO2 levels for model building</h3>

```python
Pollution = pd.DataFrame.from_dict(dict)
Pollution['Month'] = [month[int(x[3:5]) -1] for x in Pollution['Date']]
Pollution['Year'] = [x[6:10] for x in Pollution['Date']]


temp= pd.to_datetime(Pollution['Date'])
temp = temp.dt.strftime('%d-%m-%Y')
Pollution['Day'] = pd.to_datetime(temp).dt.day_name()
columns = ['City', 'Date', 'Day', 'Month', 'Year', 'Mean Pollution']
Pollution = Pollution.reindex(columns=columns)


Pollution
```
